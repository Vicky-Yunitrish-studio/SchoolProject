```python
from langchain.prompts import PromptTemplate
from langchain_ollama.llms import OllamaLLM
chat_model = OllamaLLM(model='phi3')

promptTemplate = PromptTemplate.from_template(
    "在{context}學問中 你覺得{topic}是什麼 ?"
    )

# this format the prompt template with the given parameters
question = promptTemplate.format(
    topic="愛情與麵包", 
    context="哲學")
answer = chat_model.invoke(question)
print(question)
print(answer)
```

```python
from langchain.prompts import PromptTemplate
from langchain_ollama.llms import OllamaLLM
chat_model = OllamaLLM(model='phi3')

def get_answer(topic, context):
    promptTemplate = PromptTemplate.from_template(
        "在{context}學問中 你覺得{topic}是什麼 ?"
    )
    question = promptTemplate.format(
        topic=topic, 
        context=context)
    answer = chat_model.invoke(question)
    return (question, answer)

Q, A = get_answer("愛情與麵包", "哲學")
print("User", Q)
print("Robot", A)
```

```python
from langchain.schema import  HumanMessage, SystemMessage
from langchain_ollama.llms import OllamaLLM
chat_model = OllamaLLM(model='phi3')

message = [
    SystemMessage(content="你是一諮商輔導人員"),
    HumanMessage(content="我男朋友離開我 我非常難過")
]
answer = chat_model.invoke(message)
print(answer)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
model = OllamaLLM(model='phi3')

prompt1 = ChatPromptTemplate.from_template("what is the city {person} is from?")



chain1 = prompt1 | model | StrOutputParser()

result = chain1.invoke({"person": "Biden"})
print(result)

"""
Joe Biden is from the city of Scranton, Pennsylvania.
"""
```

```python
from langchain_ollama.llms import OllamaLLM
from operator import itemgetter
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
prompt1 = ChatPromptTemplate.from_template("what is the city {person} is from?")
prompt2 = ChatPromptTemplate.from_template(
    "what country is the city {city} in? respond in {language}"
)
model = OllamaLLM(model='phi3')
chain1 = prompt1 | model | StrOutputParser()
chain2 = (
    {"city": chain1, "language": itemgetter("language")}
    | prompt2
    | model
    | StrOutputParser()
)

result = chain2.invoke({"person": "Biden", "language": "chinese"})
print(result)
```

```python
from langchain_ollama.llms import OllamaLLM
from operator import itemgetter
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate


prompt1 = ChatPromptTemplate.from_template("what is the city {person} is from?")
prompt2 = ChatPromptTemplate.from_template(
    "what country is the city {city} in? respond in {language}"
)
prompt3 = ChatPromptTemplate.from_template(
    "{country} 與台灣的關係如何呢? 請以{traditional_c}回答"
)

model = OllamaLLM(model='phi3')

chain1 = prompt1 | model | StrOutputParser()
```

```python
chain2 = (
    {"city": chain1, "language": itemgetter("language")}
    | prompt2
    | model
    | StrOutputParser()
)

chain3 = (
    {"country": chain2, "traditional_c": itemgetter("language")}
    | prompt3
    | model
    | StrOutputParser()
)

result = chain3.invoke({"person": "Biden", "language": "繁體中文"})
print(result)
```
